```python
import pandas as pd
import numpy as np
import csv
```

```python
df = pd.read_excel ('temp.xlsx')
df.head()
```

```python
idx = np.isnan(df.values).argsort(axis=1)
```


If you receive error:
typeerror: ufunc 'isnan' not supported for input types then
try replacing np.isnan with pd.isna. Pandas' isna supports category dtypes


```python
df = pd.DataFrame(
    df.values[np.arange(df.shape[0])[:, None], idx],
    index=df.index,
    columns=df.columns,
)
```

```python
df.head()
```

```python
# Reading the second data set.
df2 = pd.read_excel ('Book2.xlsx')
df2.head(3)
```

```python
df_concat = pd.concat([df2,df],axis=1)
df_concat.head(3)
```
